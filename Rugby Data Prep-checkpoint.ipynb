```python
dfeng.loc[dfeng.home_score > dfeng.away_score, 'home_points'] = 4 
dfeng.loc[dfeng.home_score < dfeng.away_score, 'home_points'] = 0
dfeng.loc[dfeng.home_score == dfeng.away_score, 'home_points'] = 2
```

```python
dfeng.head()
```

```python
dfeng.loc[dfeng.home_score < dfeng.away_score, 'away_points'] = 4 
dfeng.loc[dfeng.home_score > dfeng.away_score, 'away_points'] = 0
dfeng.loc[dfeng.home_score == dfeng.away_score, 'away_points'] = 2
```

```python
dfeng.head()
```

```python
dfeng.loc[dfeng.home_tries >= 4, 'home_try_bonus'] = 1
dfeng.loc[dfeng.home_tries < 4, 'home_try_bonus'] = 0
```

```python
dfeng.loc[dfeng.away_tries >= 4, 'away_try_bonus'] = 1
dfeng.loc[dfeng.away_tries < 4, 'away_try_bonus'] = 0
```

```python
dfeng['home_score_bonus'] = np.where((dfeng.home_score < dfeng.away_score) & ((dfeng.away_score - dfeng.home_score) <= 7), 1,0)
```

```python
dfeng['away_score_bonus'] = np.where((dfeng.home_score > dfeng.away_score) & ((dfeng.home_score - dfeng.away_score) <= 7), 1,0)
```

```python
dfeng.head()
```

```python
dfeng['home_match_points'] = dfeng['home_points']+dfeng['home_try_bonus']+dfeng['home_score_bonus']
```

```python
dfeng['away_match_points'] = dfeng.away_points+dfeng.away_try_bonus+dfeng.away_score_bonus
```
